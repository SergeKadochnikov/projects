## Задача 2
Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
query = '''
SELECT b.book_id, b.title, COUNT(rv.review_id),AVG(rt.rating)
FROM books as b
LEFT JOIN reviews rv ON rv.book_id = b.book_id
LEFT JOIN ratings rt ON rt.book_id = b.book_id
Group by b.book_id
ORDER BY COUNT(rv.review_id) DESC, AVG(rt.rating) DESC
 '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

Каждую отдельную книгу определяет её book_ID, т.к. названия могут совпадать у книг разных издательств.
С помощью такого запроса можно посмотреть самые обозреваемые книги и самые любимые (при небольшой корректировке порядка сортировки в запросе). Доступен и антирейтинг.

## Задача 3

Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
query = '''
SELECT p.publisher_id, p.publisher, COUNT(b.book_id) as num_books
FROM books as b
JOIN publishers as p ON b.publisher_id = p.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher_id, p.publisher
ORDER BY num_books DESC
LIMIT 1;

 '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

По аналогии со вторым заданием publisher_id точно определит издательство, а вместе с ним и имя.

## Задача 4

Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
query = '''
SELECT a.author, AVG(pop.avg_rating) as avg_rating
FROM 
    (
        SELECT b.author_id, r.book_id, AVG(r.rating) as avg_rating
        FROM books as b
        JOIN ratings as r ON b.book_id = r.book_id
        GROUP BY b.author_id, r.book_id
        HAVING COUNT(r.rating) >= 50
    ) as pop
JOIN authors as a ON pop.author_id = a.author_id
GROUP BY a.author_id
ORDER BY avg_rating DESC
LIMIT 1;

 '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

В первую очередь формирую таблицу с идентификаторами книг, имеющих 50 оценок и более, авторов и средней оценки книги. .
В основном запросе к полученной таблице присоединяю имена авторов и по author_id вычисляем среднее средних оценок книг автора. Сортируем по убыванию, выводим первого автора. Гари Поттер оставил след в культуре!


## Задача 5

Посчитайте среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```python
query = '''
WITH ave as 
    (
        WITH users as 
        (
            SELECT username
            FROM ratings
            GROUP BY username HAVING COUNT(rating_id) > 48
            )

        SELECT u.username, count(rev.review_id) as avarages
        FROM reviews as rev
        LEFT JOIN users as u on u.username = rev.username
        WHERE u.username IS NOT NULL
        GROUP BY u.username
        )
SELECT AVG(ave.avarages)
FROM ave;
 '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

В среднем, пользователи, поставившие 48 оценок и более постят 24 обзора на прочитанные книги.
По принципу матрешки создавали подзапросы и использовались в запросе, снова становясь подзапросом.

Сначала из списка пользователей оставил всех, кто поставил более 48 оценок.
На следующем шаге к полученным пользователям добавил сумму обзоров, сделанных каждым из указанных пользователей. Причем не все
пользователи подписываются, безымянных не посчитал. На третьем шаго берется среднее.
